```python
import pandas as pd
import numpy as np
```

```python
incidents = pd.read_excel("../data/incidents.xlsx")
incidents.columns
```

```python
incidents.shape
```

```python
drop_columns = incidents.columns[incidents.isna().sum() > 200]
incidents = incidents.drop(columns=drop_columns, axis=1)
```

```python
drop_columns = ['Классификация НС', 'Год рождения', 'Стаж работы в организации', 'Дата ликвидации происшествия', 'Время ликвидации происшествия', 'Влияние на производственный процесс (остановка производства и снижение объема добычи, переработки, остановка движения поездов, отправки авиарейсов и т.д.), силы и средства, привлеченные', 'Место происшествия', 'Область', 'Дата возникновения происшествия', 'Принятые первоочередные меры по защите персонала, локализации', 'Классификация ОМП', 'Оказание Медицинской помощи/микротравма']
incidents = incidents.drop(columns=drop_columns, axis=1)
```

```python
incidents.isna().sum()
```

```python
for x in incidents.columns:
    print(x + ":", end='\n')
    print(incidents[x].unique())
```

```python
import numpy as np

conditions = [
    incidents['Тяжесть травмы'].str.contains('летал', case=False, na=False),
    incidents['Тяжесть травмы'].str.contains('тяж', case=False, na=False)
]

values = [2, 1]

incidents['Тяжесть травмы'] = np.select(conditions, values, default=0)
```

```python
incidents['Должность пострадавшего'] = incidents['Должность пострадавшего'].fillna('Неизвестно')
incidents['Диагноз'] = incidents['Диагноз'].fillna('Неизвестно')
incidents['Пострадавшая часть тела'] = incidents['Пострадавшая часть тела'].fillna('Неизвестно')
incidents['Корректирующие меры'] = incidents['Корректирующие меры'].fillna('Нет')
incidents['Несчастный случай'] = incidents['Несчастный случай'].notna()
incidents['На рабочем месте'] = incidents['На рабочем месте'].astype(bool)
incidents['Бизнес направление'] = incidents['Бизнес направление'].fillna('Неизвестно')
incidents['Вид работ'] = incidents['Вид работ'].replace(['Другой', 'Работы не производились'], 'Other')
incidents['Вид работ'] = incidents['Вид работ'].fillna('Other')
incidents = incidents.dropna()
```

```python
incidents.to_csv("../data/incidents_upd.csv")
```
