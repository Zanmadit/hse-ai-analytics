```python
korgau = korgau.drop('Опишите ваше наблюдение/предложение', axis=1)
korgau['Фото/Документы с места выявления несоответствия или предложения'] = (
    korgau['Фото/Документы с места выявления несоответствия или предложения'].notna() & 
    (korgau['Фото/Документы с места выявления несоответствия или предложения'].astype(str).str.strip() != '')
)
```

```python
negative_types = ['Опасный фактор', 'Хорошая практика', 'Небезопасное условие ', 'Небезопасное поведение', 'Опасный случай', 'Небезопасное действие', 'Предложение (инициатива)']
korgau['риск'] = korgau['Тип наблюдения'].isin(negative_types).astype(int)
```

```python
korgau['критическое оповещание'] = (
    (korgau['риск'] == 1) & 
    (korgau['Производилась ли остановка работ?'] == False) &
    (korgau['Было ли небезопасное условие / поведение исправлено и опасность устранена?'] == False)
).astype(int)
```

```python
korgau.to_csv("../data/korgau_upd.csv")
```
